### 4번문제 정답

```python
import pandas as pd 
import numpy as np
import mglearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


#데이터셋 섞기 + 훈련세트와 테스트세트 구분
X_train, X_test, y_train, y_test = \
train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0)

#k-NN의 분류모델 사용 , knn 객체 생성 (캡슐화)
knn = KNeighborsClassifier(n_neighbors= 1)

#테스트 세트로 학습시키기
knn.fit(X_train, y_train)

#정확도 knn.score
print("테스트 세트의 정확도 {:.2f}".format(knn.score(X_test, y_test)))

#예측
X_new = np.array([[5,2.9,1,0.2]]) #반드시 2차원 ndarray로 줘야함. [[]]
print("X_new.shape:", X_new.shape)

prediction = knn.predict(X_new) #knn.predict(2차원ndarray)
print("예측:", prediction) #index로 반환함 
print("예측한 타깃의 이름:",iris_dataset['target_names'][prediction]) #ndarry[index]
```

### 5.지도학습에 대해 설명하고 분류와 회귀의 공통점과 차이점을 설명하시오.

### 5번 문제 답
- 입력과 출력 샘플 데이터가 있고, 주어진 입력으로부터 출력을 예측하고자 할 때 사용한다.
- 둘다 지도학습이며
- 분류(classification)
    - **미리정의된, 가능성 있는 여러 레이블 클래스(class label)중 하나를 예측하는 것**
    - 이진분류(binary classification)(예,아니오, 보이스피싱)와 다중분류(multi classfication)(붓꽃품종)
    - 이진분류에서 한 클래스를 양성(positive)클래스, 다른 클래스를 음성(negative)클래스라고 한다.
    - 양성 클래스 : 학습하고자하는 대상 : ex)보이스 피싱(yes)
    - 출력값 사이에 연속성이 없다. ex)영어, 프랑스.. : 영어와 프랑스 사이의 언어 : 없음.
- 회귀(regression)
    - 연속적인 숫자(=부동소수점수=실수)를 예측하는 것
    - ex)어떤 사람의 교육수준, 나이, 주거지를 바탕으로 연간소득(또는 범위)를 예측하는 것

### 6. 일반화에 대해 설명하고 과대적합, 과소적합을 이웃수와 연관지어 설명하시오.

### 6번 문제 답
- 모델이 **처음 보는 새로운 데이터**에 대해 정확하게 예측할 수 있다면 이를 훈련세트에서 테스트세트로  
일반화(generalization)되었다고 한다.
- 가능한 한 정확하게 일반화되도록 해야한다.
- 과대적합(overfitting)
    - 모델이 훈련 세트의 각 샘플에 너무 가깝게 맞춰져서 새로운 데이터에 일반화되기 어려울 때 일어난다.
- 과소적합(underfitting)
    - 반대로 너무 간단한 모델이 선택되는 것, 데이터 면면과 다양성을 잡아내지 못하고 훈련세트에도 잘 맞지 못한다.

### 8. 각 물음에 답하시오. (4개 질문)

```python
import pandas as pd 
import numpy as np
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

X , y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#1. fig, axes에 할당되는 값에 대해 설명하시오.
fig, axes = plt.subplots(1,3,figsize=(15,4))

#2. reshape(-1,1)의 의미는?, 이것은 언제 필요한가?
line = np.linspace(-3,3,1000).reshape(-1,1)

#3. for 구문 의미 
for n_neighbors, ax in zip([1,3,9],axes):
    
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c = mglearn.cm2(0), markersize = 8) 
    ax.plot(X_test, y_test, 'v', c = mglearn.cm2(1), markersize = 8) 
    
    ax.set_title("{}이웃의 훈련 스코어 : {:.2f} 테스트 스코어: {:.2f}".format(\
                            n_neighbors, reg.score(X_train, y_train),reg.score(X_test, y_test)))
    ax.set_xlabel("특성")
    ax.set_ylabel("타깃")
    
    axes[0].legend(["모델 예측", "훈련 데이터/타깃", "테스트 데이터/타깃"], loc = 'best')

#4. 그래프 해석 
```

```python
import pandas as pd 
import numpy as np
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

X , y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#1. fig, axes에 할당되는 값에 대해 설명하시오.
fig, axes = plt.subplots(1,3,figsize=(15,4))

#2. reshape(-1,1)의 의미는?
line = np.linspace(-3,3,1000).reshape(-1,1) 
#reshape(-1,1)의미 https://rfriend.tistory.com/345

#Reshape your data either using array.reshape(-1, 1) if your data has a single feature or 
#array.reshape(1, -1) if it contains a single sample.

for n_neighbors, ax in zip([1,3,9],axes):
    
    #1,3,9 이웃을 사용한 예측을 한다.
    
    #객체 생성
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    #훈련
    reg.fit(X_train, y_train)
    #
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c = mglearn.cm2(0), markersize = 8) 
    ax.plot(X_test, y_test, 'v', c = mglearn.cm2(1), markersize = 8) #c = mglearn.cm2는 그냥 색정보
    #marker = '^'와 그냥 '^'차이 :: 이은 선 아님 점만 찍냐의 차이
    
    ax.set_title("{}이웃의 훈련 스코어 : {:.2f} 테스트 스코어: {:.2f}".format(\
                            n_neighbors, reg.score(X_train, y_train),reg.score(X_test, y_test)))
    ax.set_xlabel("특성")
    ax.set_ylabel("타깃")
    
    axes[0].legend(["모델 예측", "훈련 데이터/타깃", "테스트 데이터/타깃"], loc = 'best')
    #위의 ax 순서에 따라 배치됨.
```

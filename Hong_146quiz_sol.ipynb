```python
import numpy as np

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
#타깃(weight)의 평균
mean = np.mean(perch_weight)
print(mean)
```

```python
#훈련데이터 테스트데이터 나누기 
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = \
train_test_split(perch_length, perch_weight, random_state = 42)

#input데이터를 2차원 배열로 만들어야한다.

train_input = train_input.reshape(-1,1)
test_input = test_input.reshape(-1,1)

#1차식 학습 

from sklearn.linear_model import LinearRegression
lr_1 = LinearRegression().fit(train_input, train_target)

#1차식 계수, 절편
coef1 = lr_1.coef_
intercept1 = lr_1.intercept_

#2차식 학습 데이터
train_poly = np.column_stack((train_input**2, train_input))

#2차식 학습
lr_2 = LinearRegression().fit(train_poly, train_target)

#2차식 계수, 절편 
coef2_2 = lr_2.coef_[0]
coef2_1 = lr_2.coef_[1]
intercept2 = lr_2.intercept_

#그래프를 그리자.

import matplotlib.pyplot as plt

#스타일 지정 
plt.style.use('ggplot')

#그래프 4개가 들어갈 판의 크기 설정 
fig = plt.figure(figsize=(40, 10))

#그래프 객체 생성 
ax1 = fig.add_subplot(1,4,1) #위치 1행4열 크기의 첫번째 위치
ax2 = fig.add_subplot(1,4,2)
ax3 = fig.add_subplot(1,4,3)
ax4 = fig.add_subplot(1,4,4)

#각 객체별 그래프 그리기

#ax1 : 평균 그래프

ax1.scatter(train_input, train_target)
ax1.plot([15,50], [mean,mean])
ax1.set_title("mean")

#ax2 : 거꾸로 1차식 : 평균기준 축대칭 
ax2.scatter(train_input, train_target)
ax2.plot([15,50], [2*mean - (15*coef1 + intercept1),2*mean - (50*coef1 + intercept1)])
ax2.set_title("monomial_upsidedown")

#ax3 : 1차식
ax3.scatter(train_input, train_target)
ax3.plot([15,50], [15*coef1 + intercept1,50*coef1 + intercept1])
ax3.set_title("monomial")

#ax4 : 2차식
x = np.arange(15,45)
ax4.scatter(train_input, train_target)
ax4.plot(x, coef2_2*x**2 + coef2_1*x + intercept2)
ax4.set_title("polynomial")

#각 그래프 x축 y축 label
for i in [ax1,ax2,ax3,ax4]:
  i.set_xlabel('length')
  i.set_ylabel('weight')
```

## 진응력, 진변형률 열 추가

```python
import pandas as pd
file_path = "/content/drive/MyDrive/load_extension.xlsx"
df = pd.read_excel(file_path, engine = 'openpyxl')
print(df.head())
```

```python
#열 이름 변경 
df.columns = ["lo_kN", "ex_mm", "col3", "col4", "col5"]
df.head()
```

```python
#lo_kn, ex_mm만 가져오기
df = df[["lo_kN", "ex_mm"]]
df.head()
```

```python
#공칭응력(nominal stress)열 추가

#공칭응력 = lo_kN/시편의 초기 단면적
#시편의 초기 단면적 : 122.71mm^2
#kN/mm^2 = GPa, GPa->MPa 단위변환을 위해 10^3을 해야함
df["noSt_MPa"] = (df["lo_kN"]/122.71)*1e3
df.head()
```

```python
#공칭변형률(nominal strain) 열 추가

#공칭변형률 = ex_mm / 표점거리(50mm)
#단위없음

df["noStrain"] = (df["ex_mm"]/50)
df.head(-5)
```

```python
#진응력(true stress)열 추가
#진응력 = 공칭응력*(1+공칭변형률)
df["trSt_MPa"] = df["noSt_MPa"]*(1+df["noStrain"])
df.head()
```

```python
#실수로 추가한 trStress열 삭제
#df.drop('trStress', axis = 1 , inplace = True)
df.head()
```

```python
import numpy as np
df["trStrain"] = np.log(1+df["noStrain"])
df.head(-5)
```

## df_tr변형


```python
#자연로그 진응력, 자연로그 진변형률 열 추가
df_tr["ln_stress"] = np.log(df_tr["trSt_MPa"])
df_tr["ln_strain"] = np.log(df_tr["trStrain"])
df_tr
```

```python
#진변형률^2, 진변형률 * 진응력 열 추가
df_tr["ln_strain_2"] = df_tr["ln_strain"]**2
df_tr["ln_st*str"] = df_tr["ln_strain"]*df_tr["ln_stress"]
df_tr
```

### 손계산 : 연립

```python
#계수
x = df_tr["ln_strain"].sum()
x_squared = df_tr["ln_strain_2"].sum()
xy = df_tr["ln_st*str"].sum()
y = df_tr["ln_stress"].sum()
m = 533 # 데이터 개수 
```

```python
a = np.array([[x_squared, x],[x,m]])
b = np.array([xy,y])
sol = np.linalg.solve(a,b)

sol_a = sol[0]
sol_b = sol[1]
print(sol_a, sol_b)
```

```python
#k값을 구하자.
k = np.exp(sol[1])
k
```

### 선형모델링



해당 구간을 나타낼 식을 먼저 생각  
σ = kϵ^n  
해당식을 변형,치환을 통해 선형식으로 표현가능한지 생각  
--> 양변 자연로그를 통해 가능함.  
lnσ = lnk + nlnϵ  
여기서 구해야할 파라미터는 k(절편)과 n(계수), 나머지 σ,ϵ은 input, target 데이터에 해당함.  
lnσ와 lnϵ을 각각 x y 로 치환하면
위의 식은  
y = nx + lnk다.  
식변형 과정에서 input과target모두 변형이 생겼기 때문에 input과 target 데이터에 ln을 씌운 데이터로 학습시켜야 한다.  
모델 학습으로 얻은 절편은 lnk이고 계수는 n이다.

- 위 실험에서 수학적 모델링(식)을 하는 이유는 시편간의 강도계수와 가공경화지수의 우위판단을 통해 재료 가공성과 변형의 특성을 알아보기 위함.

- 물고기문제처럼 주어진 데이터로 미지의 데이터를 추측하기 위함이 아님.

- 따라서 원본데이터로 학습하고 평가는 것이 적절하다.

n(가공경화지수): 진응력-진변형률 로그 선도에서 기울기로 n값이 클수록 네킹 발생 전까지 더 많은 
 변형이 가능하다.
